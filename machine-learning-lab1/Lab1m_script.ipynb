## 2. 数据预处理

```python
x=data2[:,0:2]
y=data2[:,[2]]
print(x.shape, x.ndim)
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:,0],x[:,1],y)
```

```python
x[:,0] = ( x[:,0] - np.min(x[:,0]))/(np.max(x[:,0])-np.min(x[:,0]))
x[:,1] = ( x[:,1] - np.min(x[:,1]))/(np.max(x[:,1])-np.min(x[:,1]))
y = ( y - np.min(y))/(np.max(y)-np.min(y))
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],y)
```

## 3. 训练线性回归模型
### 3.1 准备输入、标签、输出、学习率、模型参数

```python
#使用m存放数据点的个数
#使用np.ones创建名为x0的二维数组，规模为m*1
#使用x1存放读入数据第一列“人口数量”，存为列向量
#使用np.hstack函数将x0和x1进行合并存放于x,打印x
#使用y存放读入数据第二列“收益”，存为列向量
#使用np.zeros初始化theta,规模为2*1,打印theta

#这里的注释有误导性
m=47
x0=np.ones((m,1))
x1 = data2[:, 0].reshape(-1, 1)
x2 = data2[:, 1].reshape(-1, 1)
x = np.hstack((x0, x1))
x=np.hstack((x,x2))
print(x)

theta=np.zeros([3,1])
theta
```

### 3.2 计算代价函数

```python
#利用numpy中的函数编写代价函数
#注意避免编写一个完整的长公式，分步计算便于调试
#注意数组维度

error=np.dot(x,theta)-y
    
J = np.sum(error ** 2) / (2*m)
print(error)
print(J)
```

### 3.3 梯度下降法求模型参数

```python
#定义学习率alpha
#迭代次数iteration
#使用cost存放每次迭代后计算出的代价函数的取值
#进行梯度下降
alpha = 0.01   # alpha是学习率
iterations = 30000   # 迭代次数
cost=[]  # cost存放每次修改theta后代价函数的值

for i in range(iterations):
    error = np.dot(x,theta) - y
    J = np.sum(error ** 2) / (2*m)
    # J = np.sum(np.power((x.dot(theta)-y),2))/(2*m)
    
    grad = np.dot(x.T, error) / m
    
    theta = theta - grad * alpha
    cost.append(J)

theta
```

### 3.4 通过求导法直接计算最优解

```python
#编写公式直接求取theta最优解
theta=np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)),x.T),y)
theta
```
