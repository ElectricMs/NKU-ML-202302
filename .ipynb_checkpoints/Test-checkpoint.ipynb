```python
import requests
import time
start_problem = 393297
end_problem = 393299
base_url = "http://10.134.1.2/nkuoj/api/student/problemstate?problem=%5B{}%5D&ucTicket=16QysJw8S"
interval = 0.5 
for problem_id in range(start_problem, end_problem + 1):
    url = base_url.format(problem_id)
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print(f"问题ID: {problem_id}, 返回内容:")
            print(response.text)
        else:
            print(f"问题ID: {problem_id}, 访问失败，状态码: {response.status_code}")
    except Exception as e:
        print(f"问题ID: {problem_id}, 请求错误: {e}")
    # 确保每次尝试之间有适当的间隔
    time.sleep(interval)
print("遍历完成.")
```
